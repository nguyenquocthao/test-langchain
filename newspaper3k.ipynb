```python
import newspaper
import json

def get_text_content(url):
    article = newspaper.Article(url)
    article.download()
    article.parse()
    main_text = article.text

    return main_text

# Example usage:
# url = 'https://www.vinmec.com/vi/tim-mach/tu-van-bac-si/thong-lien-that-quanh-mang-co-nguy-hiem-khong/?link_type=related_posts'
url = 'https://python.langchain.com/docs/modules/agents/toolkits/document_comparison_toolkit'
main_text = get_text_content(url)
print(main_text)
```
